```python
import math
import numpy as np
import matplotlib.pyplot as plt

def erf(x):
	sign = 1
	if x < 0: sign = -1
	x = abs(x)
	t = 1.0 / (1.0 + 0.3275911 * x)
	return sign *(1.0 - (((((1.061405429 * t-1.453152027) * t) + 1.421413741) * t -0.28449673)* t + 0.254829592) * t * math.exp(-x * x))

def radius(al,k,z):
	p = math.sqrt(k) * math.sin(al)
	return (p * math.exp(p * p) * erf(p) * math.sqrt(math.pi)+ 1) * 2- z *math.tan(al)

def passed(al,k,z,rd):
	if np.abs(radius(al,k,z))<rd: return 1
	return 0

def App(als,ks,z,rd):
  L=[]
  for k in ks:
    count=0
    for a in als:count+=passed(a,k,z,rd)
    L.append(count)
  return L

z=6.1298
ks=np.arange(1.3065, 1.3120,0.0001)
als=np.arange(0.6686,0.8082,(0.8082-0.6686)/99)
rd=0.00542

# Plotting a figure of width 9 and height 4
plt_1 = plt.figure(figsize=(9.8, 4))

T=App(als,ks,z,rd)
plt.plot(ks,T)
plt.show()


# Plotting a figure of width 12 and height 4
plt_1 = plt.figure(figsize=(3, 4))

z=6.1643 
als=np.arange(0.6705,0.8100,(0.8100-0.6705)/99)
rd=0.00139
ks=np.arange(1.3171,1.3187,0.00005) 
T=App(als,ks,z,rd)

plt.plot(ks,T)
plt.show()
```
